```python
df["year_built"]=df["year_built"].fillna(df["year_built"].median())
df.head()
```

```python
df["year_built"]=df["year_built"].astype(int)
df.head()
```

```python
df.dropna(inplace=True)
df.head(12)
```

```python
df["beds"]=df["beds"].fillna(0,inplace=True).astype(int)
df["baths"]=df["baths"].fillna(0,inplace=True).astype(int)
df.head()
```

```python
df["garage"]=df["garage"].fillna(0,inplace=True)
df["garage"]=df["garage"].astype(int)
df.head()
```
